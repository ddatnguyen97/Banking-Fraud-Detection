```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train1 = pd.read_csv('Fraud Data/Train Data/fraudTrain_part1.csv')
train2 = pd.read_csv('Fraud Data/Train Data/fraudTrain_part2.csv')
train3 = pd.read_csv('Fraud Data/Train Data/fraudTrain_part3.csv')
train4 = pd.read_csv('Fraud Data/Train Data/fraudTrain_part4.csv')
train5 = pd.read_csv('Fraud Data/Train Data/fraudTrain_part5.csv')
train_df = pd.concat([train1, train2, train3, train4, train5])
```

```python
pd.options.display.float_format = '{:.2f}'.format
train_df.describe()
```

```python
plt.figure(figsize=(10, 8))
plot = sns.countplot(x='is_fraud', data=train_df)
y_max = plt.gca().get_ylim()[1]

for p in plot.patches:
    plt.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height() + y_max * 0.02),
                 ha = 'center', va = 'center')
    
plt.title('Count of Fraudulent vs. Non-Fraudulent Transactions')
plt.legend(['Not Fraud', 'Fraud'])
```

```python
donut = train_df['is_fraud'].value_counts()

plt.pie(donut, autopct = '%1.1f%%', startangle = 90, colors = ['#2bb179', '#ff000d'], pctdistance=0.8,
        explode=(0,0))
centre_circle = plt.Circle((0.0, 0.0), 0.5, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Percentage of Fraudulent vs. Non-Fraudulent Transactions')
plt.legend(['Not Fraud', 'Fraud'])
```

```python
plt.figure(figsize=(10, 8))
plot = sns.histplot(train_df['amt'], kde=True)
```

```python
train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])
train_df['trans_date'] = train_df['trans_date_trans_time'].dt.date
train_df['trans_time'] = train_df['trans_date_trans_time'].dt.time
train_df['year'] = train_df['trans_date_trans_time'].dt.year
train_df['quarter'] = train_df['trans_date_trans_time'].dt.quarter
train_df['month'] = train_df['trans_date_trans_time'].dt.month
train_df['day'] = train_df['trans_date_trans_time'].dt.day
```

```python
train_df = train_df.drop(columns=['trans_date_trans_time', 'Unnamed: 0',])
train_df.head()
```

```python
plt.figure(figsize=(10, 8))
plot = sns.countplot(x='year', data=train_df)
y_max = plt.gca().get_ylim()[1]

for p in plot.patches:
    plt.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height() + y_max * 0.02),
                 ha = 'center', va = 'center')
plt.title('Count of Transactions by Year')
```
